```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset containing paths to real and fake logos
real_logos_df = pd.read_csv('D:/pythn_prj/New/logo_dataset/real_logo.csv')
fake_logos_df = pd.read_csv('D:/pythn_prj/New/logo_dataset/fake_logo.csv')

# Count the number of real and fake logos for each brand
real_logos_count = real_logos_df['Brandname'].value_counts()
fake_logos_count = fake_logos_df['Brand Name'].value_counts()

# Merge the counts into a single DataFrame
logo_counts_df = pd.DataFrame({'Brandname': real_logos_count.index, 'Real Logos': real_logos_count.values})
logo_counts_df['Fake Logos'] = logo_counts_df['Brandname'].map(fake_logos_count)

# Fill missing values with 0
logo_counts_df.fillna(0, inplace=True)

# Plotting bar chart

logo_counts_df.plot(x='Brandname', kind='bar', figsize=(12, 6))
plt.xlabel('Brand Name')
plt.ylabel('Number of Logos')
plt.title('Number of Real and Fake Logos for Each Brand')
plt.xticks(rotation=90)
plt.legend(['Real Logos', 'Fake Logos'])
plt.show()
```

```python
# Load your dataset
data = pd.read_csv('D:/pythn_prj/New_project/combined.csv')

# Remove rows with missing labels
data = data.dropna(subset=['Label'])

# Visualize label distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Create a subplot with 1 row and 2 columns, and activate the first subplot
data['Label'].value_counts().plot(kind='bar')
plt.title('Label Distribution')
plt.ylabel('Count')
 
plt.subplot(1, 2, 2)  # Activate the second subplot
plt.pie([logo_counts_df['Real Logos'].sum(), logo_counts_df['Fake Logos'].sum()], labels=['Real Logos', 'Fake Logos'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Real and Fake Logos')
plt.axis('equal')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Plotting scatter plot
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)  # Create a subplot with 1 row and 2 columns, and activate the first subplot
plt.scatter(logo_counts_df['Brandname'], logo_counts_df['Real Logos'], color='blue', label='Real Logos')
plt.scatter(logo_counts_df['Brandname'], logo_counts_df['Fake Logos'], color='red', label='Fake Logos')
plt.xlabel('Brand Name')
plt.ylabel('Number of Logos')
plt.title('Scatter Plot of Real and Fake Logos by Brand Name')
plt.legend()
plt.xticks(rotation=90)
plt.grid(True)

# Compute the total counts for real and fake logos
total_real_logos = real_logos_df.notnull().sum()
total_fake_logos = fake_logos_df.notnull().sum()

# Combine the counts into a single DataFrame
combined_heatmap_data = pd.DataFrame({'Real Logos': total_real_logos, 'Fake Logos': total_fake_logos})

# Plot the combined heatmap
plt.subplot(1, 2, 2)  # Activate the second subplot
sns.heatmap(combined_heatmap_data, cmap='coolwarm', annot=True, fmt=".1f")  # Use ".1f" for floats with one decimal place
plt.title('Distribution of Real and Fake Logos')
plt.ylabel('Features')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
